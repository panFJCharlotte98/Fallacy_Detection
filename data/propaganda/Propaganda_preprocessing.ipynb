```python
import nltk
# nltk.download('punkt')
import nltk.data
import json
import re
from os import listdir
from os.path import isfile, join

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

# file paths
dataset_path = "./data/Propanganda/datasets-v2/datasets"
train_article_path = dataset_path + '/train-articles/'
train_labels_path = dataset_path + "/train-labels-FLC/"
train_articles = [train_article_path + f for f in listdir(train_article_path) if isfile(join(train_article_path, f))]
train_labels = [train_labels_path + f for f in listdir(train_labels_path) if isfile(join(train_labels_path, f))]

# print(open(test_name, 'r', encoding='utf-8').read())
label_dict = {
        'AdHominem': 'Ad Hominem', 
        'AppealtoEmotion': 'Appeal to Emotion', 
        'AppealtoAuthority': 'Appeal to Authority', 
        'Slogans': 'Slogans', 
        'Slipperyslope': 'Slippery Slope', 
        'FalseCause': 'False Cause',
        'Slogans': 'Slogans',
        'Black-and-White_Fallacy': 'Black-and-White Fallacy',
        'Loaded_Language': 'Loaded Language',
        'Flag-Waving': 'Flag-Waving',
        'Name_Calling': 'Name Calling',
        'Labeling': 'Labeling',
        'Whataboutism': 'Whataboutism',
        'Causal_Oversimplification': 'Causal Oversimplification',
        'Exaggeration': 'Exaggeration',
        'Minimisation': 'Minimisation',
        'Doubt': 'Doubt',
        'Appeal_to_Authority': 'Appeal to Authority',
        'Repetition': 'Repetition',
        'Appeal_to_fear-prejudice': 'Appeal to fear-prejudice',
        'Red_Herring': 'Red Herring',
        'Thought-terminating_Cliches': 'Thought-terminating Cliches',
        'Bandwagon': 'Bandwagon',
        'Reductio_ad_hitlerum': 'Reductio ad hitlerum',
        'Obfuscation': 'Obfuscation',
        'Intentional_Vagueness': 'Intentional Vagueness',
        'Confusion': 'Confusion',
        'Straw_Men': 'Straw Men'
}
```

```python
def article_to_fallacies(article_path, labels_path, sentences):
    article_file = open(article_path, 'r', encoding="utf-8")

    # the tokenizer often confuses the title and first sentence to be one, so we must split them
    
    article_title = article_file.readline()
    # print(article_title, article_path)
    article = article_file.read()
    article_sents = tokenizer.tokenize(article)
    article = article_title + article
    article_sents.insert(0, article_title)

    already_appeared = []
    for line in open(labels_path, 'r', encoding="utf-8"):
        
        _, labels, start, end = line.split()
        labels = labels.split(',')
        for i in range(len(labels)):
            labels[i] = label_dict[labels[i]]
            
        fallacious_text = article[int(start) : int(end)-1]
        # print(fallacious_text)

        is_found = False
        for i in range(len(article_sents)):
            if fallacious_text in article_sents[i]:
                is_found = True
                break

        # print(i)
        if is_found and article_sents[i] not in already_appeared:
            already_appeared.append(article_sents[i])
            sentences.append({
                "id": len(sentences),
                "title": article_title,
                "pre_text": "" if i == 0 else article_sents[i-1] if i == 1 else article_sents[i-2] + article_sents[i-1],
                "text": article_sents[i],
                "post_text": "" if i == len(article_sents)-1 else article_sents[i+1] if i == len(article_sents)-2 else article_sents[i+1] + article_sents[i+2],
                "labels": labels,
            })

        # print repeated sentence
        # elif article_sents[i] in already_appeared:
        #     print(article_sents[i])

    return(sentences)
```

```python
sentences = []

for i in range(len(train_articles)):
    sentences = article_to_fallacies(train_articles[i], train_labels[i], sentences)

Propaganda_out = open("Propaganda.json", "w")
Propaganda_out.write(json.dumps(sentences, indent=4))
Propaganda_out.close()
```
